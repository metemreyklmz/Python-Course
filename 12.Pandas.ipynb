```python
# Pandas Module: This is a commonly used module for working with data

import pandas as pd
import numpy as np
```

```python
# The idea in pandas is to store data in a table
# That is easy to work with and easy to look at and easy to extracy information

var1 = np.random.randn(100)*5 + 20
# Generates a random number from a data distribution
# it can be beta, binominal, chi square distribution
# randn creates a normal distribution or Gaussian Distribution
print(var1)
print(np.mean(var1))
```

```python
var2 = np.random.randn(100)>0
var2
```

```python
labels = ['Temp. (C)', 'Ice cream eaten']
# The is simulating temperature with var1
# Simulating ice cream eaten as var2. True means eaten and False means haven't eaten

# create a dictionary that contains the dataset

D = {labels[0]:var1,labels[1]:var2}
D # it is not a good way to examine this dictionary
```

```python
# import the dictionary into a pandas dataframe
df = pd.DataFrame(data=D) # this is a different way to give an input.
# You can give this type of input when input or order of input is optional.
df

```

```python
df.head(10) # it returns first couple of rows
# you can give an input like 10. If you leave it empty, it will give a couple of rows
```
