```python
def fz():
  print(z)
```

```python
fz() # it will give an error because we don't define z inside the function or anywhere
```

```python
z = 3 # if we define z here and run, it will give you the result
fz()
```

```python
# Rules:
# 1) variables created inside a function are local (accessible only inside the function)
# 2) variables created outside a function are global (accessible in or out of the function)
# local variables have priority over global variables
```
