```python
# multiple outputs

def funfun(in1,in2):
  sumres  = in1 + in2
  prodres = in1 * in2
  print('Their sum is ' + str(sumres))
  print('Their product is ' + str(prodres))
  return sumres, prodres
```

```python
out = funfun(12,2) # output of this function is a tuple

print(type(out))
print(out[1])
```

```python
outS, outP = funfun(4,5)
print(outS, outP)
```

```python
# Exercise1: Create a function that computes a factorial of an integer

# 5! = 5*4*3*2*1
# Exercise2: there is a module named math, find the factorial function inside this module
# compare you result with this function
```

```python
import numpy as np
import math
```

```python
val = 4

print(np.arange(val)) # it will give you an array from 0 to 3
print(np.prod(np.arange(val))) # it will give you zero as output
```

```python
# but we want it starts from 1 end with 4
np.prod(np.arange(1,val+1))
```

```python
def myfactorial(val):
  return np.prod(np.arange(1,val+1))
```

```python
myfactorial(4)
```

```python
math.factorial(4)
```

```python
k = 23
# k = 23 # it will give a precision mistake. The out will be huge number
print(myfactorial(k),math.factorial(k))

```
