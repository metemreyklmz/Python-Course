```python
# if-else
# It is a flow control statement
# Flow control statements are a way to allow the computer to make desicions and run the same code multiple times depending on the state of particular variables

if True: # if the following condition is True (in this case is 'True'), do the following statements
  print('True is true')
```

```python
if False:
  print('False is true')
```

```python
r = np.random.randn()
if r > 0:
  print(str(r) + ' is positive')
else: # if r>0 is false the else statement runs
  print(str(r) + ' is negative')
```
