```python
# Classes and Object Oriented Programming(OOP)
# Classes are pretty common in Python
# Even if you do not create a clasess, it is useful to know
# It is very similar to a dictionary. Also it has specific functions that are known as methods
# This methods are only related to this object

class Experiment(object):
  # constructor method: is basically how the object created or how we create this variable
  def __init__(self,eType,dataN,status='ongoing'): # first input is self(itself of the class).
  # the others are features of this class. eType means type of experiment.
  # status is another feature. and its default valur is ongoing
  # these inputs are needed. but status is optional because it has default value
    self.expType = eType # like in a dictionary expType is key and eType is the value
    self.dataCount = dataN # if you want you can give to key same name as the value. Like selfdataN = dataN
    self.expStatus = status

  def increaseN(self,number):
    self.dataCount += number

  def changeExpStatus(self,status):
    self.expStatus = status

  def __str__(self):
    return 'Experiment type: ' + self.expType + ' has ' + str(self.dataCount) + ' data points.'
```

```python
# Let's create a new experiment from this class

myexperiment = Experiment('Brain',120) # this codes initialize the object. 120 is data points

myexperiment
```

```python
myexperiment.dataCount
```

```python
print(myexperiment.dataCount)
myexperiment.increaseN(10)
print(myexperiment.dataCount)
```

```python
print(myexperiment)
```
