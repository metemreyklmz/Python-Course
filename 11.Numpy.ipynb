```python
n = [4,5,6,7,1,2,6,4,8]
print(n)
n.sort() # some the methods change the variable like sort method
# from now on n is sorted and remains like that
print(n)
```

```python
# some methods can change the variable type
n = [4,5,6,7,1,2,6,4,8]
print(type(n))
n = np.sort(n)
print(type(n))
```

```python
# you have to understand that n.sort is a method and modifies the list(or numpy array) in place
# and doesn't give a result
# but np.sort() is Numpy function, not a method
# it doesn't modify the input array
# it returns a new sorted copy
# If you don't want to keep original array use .sort() method
# If you want to keep original, use np.sort(n) function
```
